```python
import pandas as pd

# Direct link to the raw Excel file on GitHub
url = 'https://github.com/Rising-Stars-by-Sunshine/STATS201_Jiahe_PS2/raw/main/data/test%20data.xlsx'

# Read the Excel file into a DataFrame
df2 = pd.read_excel(url, engine='openpyxl', index_col=None)

# Display the first few rows of the DataFrame
df2.head()
```
