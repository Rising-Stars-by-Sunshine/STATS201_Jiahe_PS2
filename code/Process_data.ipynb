```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# Read the Excel file from the URL
url = 'https://github.com/Rising-Stars-by-Sunshine/STATS201_Jiahe_PS2/raw/main/data/cancer%20patient%20data%20sets%20(1).xlsx'
df = pd.read_excel(url, engine='openpyxl', index_col=None)

# Display the first few rows of the DataFrame
df.head()

```

```python
# Define a mapping dictionary for the levels
level_mapping = {'Low': 0, 'Medium': 1, 'High': 2}

# Map the levels to numerical values
df['Target'] = df['Level'].map(level_mapping)

# Display the first few rows of the DataFrame
print(df.head())
```
